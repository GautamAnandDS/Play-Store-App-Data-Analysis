# **Project Summary -**

The rapid expansion of the mobile app market has made it imperative for developers, marketers, and business strategists to understand the landscape of applications available on platforms like the Google Play Store. This project focused on an Exploratory Data Analysis (EDA) of Google Play Store app data to extract meaningful insights and trends that can inform various stakeholders in the mobile app ecosystem. The dataset used for this analysis includes various attributes such as 'App', 'Category', 'Rating', 'Reviews', 'Size', 'Installs', 'Type', 'Price', 'Content Rating', 'Genres', 'Last Updated', 'Current Ver', and 'Android Ver'.

Objectives
The primary objectives of this EDA project were to:

Understand the Distribution and Composition of the Play Store: Analyze the categories of apps available, the distribution of ratings, and the prevalence of free vs. paid apps.
Identify Key Trends in User Engagement: Examine metrics like app ratings, reviews, and installs to understand how different factors contribute to an app's popularity and user satisfaction.
Explore the Impact of App Characteristics on Success: Investigate how features such as app size, price, and content rating affect user engagement and app performance.
Provide Actionable Insights for Stakeholders: Generate insights that can help developers optimize their app offerings, marketers target the right audience, and business strategists make informed decisions.
Data Overview
The dataset comprises multiple features that provide a comprehensive overview of each app's characteristics and performance metrics:

* App: The name of the application.
* Category: The category under which the app is listed, such as 'Tools', 'Games', 'Finance', etc.
* Rating: The average rating given by users, ranging from 1 to 5.
* Reviews: The number of user reviews the app has received.
* Size: The size of the app in megabytes.
* Installs: The number of times the app has been installed by users.
* Type: Whether the app is 'Free' or 'Paid'.
* Price: The cost of the app, applicable only for paid apps.
* Content Rating: The suitability of the app content for different age groups, such as 'Everyone', 'Teen', etc.
* Genres: A more granular classification of the app category.
* Last Updated: The date when the app was last updated.
* Current Ver: The current version of the app.
* Android Ver: The minimum Android version required to run the app.

Changing datatypes to integer or float according to the requirements with some manipulation

````python
# Changing datatype of reviews to integer
df['Reviews'] = df['Reviews'].astype(int)
````

````python
# changing datatype of size to float with some manipulation
#creating a custom function for this
def convert_size(size):
    if 'M' in size:
        return float(size.replace('M', ''))
    elif 'k' in size:
        return float(size.replace('k', '')) / 1024
    else:
        return size

# Apply the custom function to the 'Size' column
df['Size'] = df['Size'].apply(convert_size)
# Converting the rows with "varies with devices" to NaN
df['Size'] = pd.to_numeric(df['Size'], errors='coerce')
````

````python
# Converting the installs column to numeric
# Creating a custom function for this
def convert_installs(installs):
    if '+' and ',' in installs:
        return int(installs.replace('+', '').replace(',', ''))
    elif '+' in installs:
        return int(installs.replace('+', ''))
    else:
        return installs

# Apply the custom function to the 'Installs' column
df['Installs'] = df['Installs'].apply(convert_installs)
df['Installs']=df['Installs'].astype(int)
````

````python
# Converting price column to float
# defining a custom function
def convert_price(price):
    if price == '0':
        return 0
    elif "$" in price:
        return float(price.replace('$', ''))
    else:
        return price

# Apply the custom function to the 'Price' column
df['Price'] = df['Price'].apply(convert_price)
df['Price']=df['Price'].astype(float)
````

````python
# Converting last updated to datetime format
df['Last Updated'] = pd.to_datetime(df['Last Updated'])
````

# **categorizing apps by rating in 5, i.e-(1.0 to 2),(2.1 to 3.0)...(4.1 to 5).**

````python
# Define the rating bins
rating_bins = [1, 2.1, 3.1, 4.1, 5.1]
# Define the rating labels
rating_labels = ['1.0 to 2.0', '2.1 to 3.0', '3.1 to 4.0', '4.1 to 5.0']

# Categorize the ratings and store the result in a new variable
rating_Rcategories = pd.cut(df['Rating'], bins=rating_bins, labels=rating_labels, right=False)
# Count the number of apps in each rating category
category_Rcounts = rating_Rcategories.value_counts()
# Print the counts
print(category_Rcounts)
````

# **categorizing apps by installs**

````python
# Define install bins
install_bins=[0,250000001,500000001,750000001,10000000001]
# Define install lebels
install_labels = ['0 to 250000000', '250000001 to 500000000', '500000001 to 750000000', '750000001 to 10000000000']
# Categorize the install and store the result in a new variable
install_categories=pd.cut(df['Installs'],bins=install_bins,labels=install_labels,right=False)
# Count the number of apps in each install category
install_counts=install_categories.value_counts(sort=False)
# Print the counts
print(install_counts)
````

## **categorizing apps by price**

````python
# Define price bins
price_bins=[0.1,1.1,10.1,20.1,50.1,100.1,451]
# Define price labels
price_lebels=['0.1 to  1.0','1.1 to 10.0','10.1 to 20.0','20.1 to 50.0','50.1 to 100.0','100.1 and more']
# Categorize the price and store the result in a new variable
price_categories=pd.cut(df['Price'],bins=price_bins,labels=price_lebels,right=False)
# Count the number of apps in each price category
price_counts=price_categories.value_counts(sort=False)
# Print the counts
print(price_counts)
````

````python
# Chart - 6 visualization code
# Count the number of occurrences for each genre and sort by count
genre_counts = df['Genres'].value_counts().sort_values(ascending=False)

# Number of genres per subplot
genres_per_subplot = int(np.ceil(len(genre_counts) / 12))

# Split the genres into 12 groups for plotting
splits = [genre_counts[i:i + genres_per_subplot] for i in range(0, len(genre_counts), genres_per_subplot)]

# Ensure there are 12 groups, filling with empty data if necessary
while len(splits) < 12:
  splits.append(pd.Series(dtype='int'))

# Plot the data in 4 rows and 3 columns using bar plots
fig, axes = plt.subplots(4, 3, figsize=(18, 24), sharey=True)

# Flatten axes for easy iteration
axes = axes.flatten()

for i, subset in enumerate(splits):
  bars = axes[i].bar(subset.index, subset.values, color='lightgreen', edgecolor='black')
  axes[i].set_title(f'Plot {i+1}')
  axes[i].set_xticklabels(subset.index, rotation=90)

    # Add value labels on the bars
  for bar in bars:
    yval = bar.get_height()
    axes[i].text(bar.get_x() + bar.get_width()/2, yval + 10, int(yval), ha='center', va='bottom')

  if i % 3 == 0:
    axes[i].set_ylabel('Count')

axes[i].set_ylim(0, 900)  # Adjust based on your data

# Final adjustments and showing the plot
plt.tight_layout()
# set background color in subplot
for ax in axes:
  ax.set_facecolor('lightpink')
plt.show()
````
